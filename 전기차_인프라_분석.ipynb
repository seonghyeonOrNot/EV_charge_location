```python
# 주차 공간 면적 계산
df['Parking Area'] = df['Number of stories/generations'] * 12.5

# 컬럼 순서 변경
dff = df[['Apartment', 'Address', 'Number of units', 'Number of floors', 'Number of stories/generations', 'Parking Area', 'Move-in year', 'Latitude', 'Longitude']]

# 결과 출력
dff
```

```python
# 충전기기 설치수 계산 함수
def calculate_charging_stations(row):
    if row['Move-in year'] <= 2021:
        return row['Parking Area'] * 0.02
    else:
        return row['Parking Area'] * 0.05

# 충전기기 설치수 컬럼 추가
dff['Number of charging device'] = dff.apply(calculate_charging_stations, axis=1)

# 결과 출력
dff
```

```python

# 'Number of charging device' 컬럼 위치 변경
dff = dff[['Apartment', 'Address', 'Number of units', 'Number of floors', 'Number of stories/generations', 'Move-in year', 'Parking Area', 'Number of charging device','Latitude', 'Longitude' ]]

# 'Number of charging device' 컬럼 값을 올림하여 정수로 변환
dff['Number of charging device'] = np.ceil(dff['Number of charging device']).astype(int)

dff
```
